```python
L = ["Ram", 1, "Shyam", 2, "Aman", 3]
print("Original list:", L)

# Separate and sort numbers and strings, then join them
L = sorted([x for x in L if type(x) == int]) + sorted([x for x in L if type(x) == str])
print("Sorted list:", L)
```
